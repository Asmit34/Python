```python
df.rename(columns={
    "Runtime (Minutes)":"Runtime",
    "Revenue (Millions)":"Revenue"
},inplace=True)
df
```

```python
#df.columns=[col.lower() for col in df.columns]
df.columns=[col.lower() for col in df.columns]
df
```

```python
df.isna().sum()
```

```python
revenue_mean=df["revenue"].mean()
revenue=df["revenue"]
```

```python
revenue.fillna(revenue_mean,inplace=True)
df.revenue
```

```python
df.isna().sum()
```

```python
metascore_mean=df["metascore"].mean()
metascore=df["metascore"]
```

```python
metascore.fillna(metascore_mean,inplace=True)
df.metascore
```

```python
df.isna().sum()
```

```python
def check_revenue(x):
    if x>revenue_mean:
        return "Blockbuster"
    else:
        return "Flop"
```

```python
data=df["revenue"].apply(check_revenue)
```

```python
df["performance"]=data
df
```
