```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3,4])
b=np.array([2,3,4,1])
plt.subplot(2,1,1)
plt.plot(a,b)
plt.title("plot1")
plt.show()

c=np.array([1,2,3,4])
d=np.array([1,2,3,4])
plt.subplot(2,1,2)
plt.plot(c,d)
plt.title("plot2")
plt.show()#-->for two different graph

plt.subplots_adjust(left=0.2,right=0.4,top=0.6,bottom=0.3)
```

```python
plt.plot(a,b,"*--g",ms=10)
```

# SCATTER PLOT

```python
x=np.array([2,3,2,3,4,6,7,6,9])
y=np.array([8,7,6,5,9,3,2,1,4])
plt.subplot(2,1,1)
plt.scatter(x,y,alpha=0.9,color="green")
plt.title("plot1")

x1=np.array([1,2,3,4,5,4,3,2,2])
x2=np.array([1,2,3,7,6,5,4,7,3])
plt.subplot(2,1,2)
plt.scatter(x1,x2,c="grey")
plt.title("plot2")
```

# Bar Graph

```python
Company=["GOOGLE","AMAZON","APPLE","NETFLIX"]
Employee=[1100,2400,3400,6000]
Branch=[1000,2000,1400,1100]

index=np.arange(len(Company))
plt.xticks(index,Company)
plt.subplot(2,1,1)
plt.barh(Company,Employee,label="Employee")
plt.subplot(2,1,2)

plt.barh(Company,Branch,label="Branch",)


#plt.barh(Company,Employee)
plt.legend()
```

```python
plt.barh(Company,Employee,height=0.8)
```
