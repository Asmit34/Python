```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import style
```

```python
x=np.array([1,2,3,4,5,6])
y=np.array([6,5,4,3,2,1])
plt.subplot(2,1,1)
plt.plot(x,y,"o--b",ms=10)
plt.title("Graph1")

plt.show()


x1=np.array([3,6,7,8,9,3])
y1=np.array([5,4,3,5,7,8])
plt.subplot(2,1,2)
plt.plot(x1,y1,"<--r",ms=10)
plt.title("Graph2")

plt.show()
```

# Scattter plot

```python
plt.figure(figsize=(16,9))
a=np.array([4,5,3,7,9])
b=np.array([3,8,6,3,3])
plt.subplot(2,1,1)
plt.title("Plot1")
plt.scatter(a,b,c="g",alpha=1,marker=r'$\clubsuit$',s=1118)

c=np.array([8,4,7,4,1])
d=np.array([7,9,2,3,5])
plt.subplot(2,1,2)
plt.title("PLot2")
plt.scatter(c,d,c="b",alpha=0.6)
```

# Bar Graph

```python
Classes=["Pythom","JAVA","Frontend","Larabel"]
Class_Students1=[10,14,16,17]
Class_Students2=[12,11,15,10]
Class_Students3=[12,22,23,7]
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(Classes,Class_Students1,width=0.6,align="edge",alpha=0.9,color="k",edgecolor="y",linewidth=4,linestyle="--",label="Class 1 Students")
plt.title("Bar Graph of Different classes",fontsize=19)
plt.xlabel("No. of student",fontsize=19)
plt.ylabel("No. of subject",fontsize=19)
#visible=False
```

```python

Classes_index=np.arange(len(Classes))
plt.barh(Classes,Class_Students1)
```

```python
plt.figure(figsize=(16,9))
plt.barh(Classes,Class_Students1,height=0.6,align="edge",alpha=0.9,color="k",edgecolor="y",linewidth=4,linestyle="--",label="Class 1 Students")
plt.title("ww")
```

```python

plt.figure(figsize=(16,9))
width=0.2

Classes_index=np.arange(len(Classes))

plt.bar(Classes_index,Class_Students1,width,color="b",label="Class 1 Students")

plt.bar(Classes_index + width,Class_Students2,width,color="y",label="Class 2 Students")

plt.bar(Classes_index + width+width,Class_Students3,width,color="g",label="Class 3 Students")  
plt.xticks(Classes_index + width,Classes,rotation=20)
plt.title("Bar Graph of Different classes",fontsize=19)
plt.xlabel("No. of student",fontsize=19)
plt.ylabel("No. of subject",fontsize=19)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
width=0.2

Classes_index=np.arange(len(Classes))

plt.barh(Classes_index,Class_Students1,width,color="b",label="Class 1 Students")

plt.barh(Classes_index + width,Class_Students2,width,color="y",label="Class 2 Students")\

plt.barh(Classes_index + width+width,Class_Students3,width,color="g",label="Class 3 Students")  
plt.xticks(Classes_index + width,Classes,rotation=20)
plt.title("Bar Graph of Different classes",fontsize=19)
plt.xlabel("No. of student",fontsize=19)
plt.ylabel("No. of subject",fontsize=19)
plt.legend()
plt.show()
```
